```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline

pf=pd.read_excel("./pf.xlsx")
f=pd.read_excel("r.xlsx")
for i in pf.columns[1:]:
    f['pred_'+i]=pf[i]
f.set_index('Date', inplace=True)
```

```python
f[['pred_factor5','factor5']].plot(figsize=(12,5));
```

```python
for i in pf.columns[1:]:
    f['re_'+i]=((f['pred_'+i]-f[i])/(f['factor1']-f['factor1'].shift(1))).apply(lambda x: abs(x))
    
f[f.columns[-5:]].iloc[-132:].median()
```
